```python
#smoking
smoking_by_gender = pd.crosstab(df_Final['Gender'], df_Final['Smoking'])
print(smoking_by_gender)
#so percentage of male smoker is more than female so filling  yes in column
#smoking where the gender is male
```

```python
# Fill missing Smoking values based on Gender
"""#so percentage of male smoker is more than female so filling  yes in column
#smoking where the gender is male"""
df_Final.loc[(df_Final['Smoking'].isnull()) & (df_Final['Gender'] == 'Male'), 'Smoking'] = 'Yes'
df_Final.loc[(df_Final['Smoking'].isnull()) & (df_Final['Gender'] == 'Female'), 'Smoking'] = 'No'
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df_t=pd.read_csv("/content/drive/MyDrive/Dataset/Titanic-Dataset.csv")

```

```python
df_t.head(50)
```

```python

first_class_df = df_t.loc[df_t['Pclass'] == 1, ['Name', 'Pclass', 'Sex', 'Age', 'Fare', 'Survived']]


fare_mean = first_class_df['Fare'].mean()
fare_median = first_class_df['Fare'].median()
fare_max = first_class_df['Fare'].max()
fare_min = first_class_df['Fare'].min()

print(f"Mean Fare: {fare_mean}")
print(f"Median Fare: {fare_median}")
print(f"Max Fare: {fare_max}")
print(f"Min Fare: {fare_min}")
```

```python
nullAge = first_class_df['Age'].isnull().sum()
print(f"Number of null values in Age: {nullAge}")

# Drop rows with null Age
first_class_df = first_class_df.dropna(subset=['Age'])
```

```python
first_class_df.isnull().sum()
```

```python
print(df_t['Embarked'].unique())

embarked_dummies = pd.get_dummies(df_t['Embarked'], prefix='Embarked')
for col in embarked_dummies.columns:
    df_t.loc[:, col] = embarked_dummies[col]
df_t = df_t.drop('Embarked', axis=1)
df_t.head()
```

```python
survival_by_gender = df_t.loc[:, ['Sex', 'Survived']].groupby('Sex').mean()
print(survival_by_gender)
```

```python
survival_by_sex_embarked = df_t.loc[:, ['Sex', 'Embarked_C', 'Embarked_Q', 'Embarked_S', 'Survived']]
print(survival_by_sex_embarked)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Mean survival for each port using the one-hot columns
survival_C = df_t.loc[df_t['Embarked_C'] == 1, 'Survived'].mean()
survival_Q = df_t.loc[df_t['Embarked_Q'] == 1, 'Survived'].mean()
survival_S = df_t.loc[df_t['Embarked_S'] == 1, 'Survived'].mean()

# Create a summary DataFrame
survival_by_port = pd.DataFrame({
    'Port': ['C', 'Q', 'S'],
    'SurvivalRate': [survival_C, survival_Q, survival_S]
})

print(survival_by_port)

```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(6,4))
sns.barplot(data=survival_by_port, x='Port', y='SurvivalRate', palette='Set2')
plt.title('Survival Rate by Port of Embarkation')
plt.ylabel('Mean Survival Rate')
plt.ylim(0,1)
plt.show()
```
