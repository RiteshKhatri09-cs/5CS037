```python
#linear algebra with matrix
import numpy as n
""" 2x − 3y + z = −1
    x − y + 2z = −3
    3x + y − z = 9   """

a=n.array([[2, -3, 1],
          [1, -1, 2],
          [3, 1, -1]],dtype=float);

b=n.array([-1, -3, 9],dtype=float);


aInverse=n.linalg.inv(a)

solution=n.dot(aInverse, b);
x=solution[0]
y=solution[1]
z=solution[2]
print(f"x:{x}, y:{y}, z:{z}")



```
