```python

df['Age_Group'] = ''

df.loc[df['Age'] < 18, 'Age_Group'] = 'child'
df.loc[(df['Age'] >= 18) & (df['Age'] <= 64), 'Age_Group'] = 'adult'
df.loc[df['Age'] >= 65, 'Age_Group'] = 'senior'

```

```python
#finding total number of passengers
# the number of passenger is equal to the number of the row in the data set
shape = df.shape
print(f"The total number is passenger  is :{shape[0]}")
# the total number is
```
